```python
import pandas as pd
import matplotlib.pyplot as plt

#lecture  CSV
df = pd.read_csv("resultats.csv")

#affichage
display(df)

# courbes
plt.figure(figsize=(10, 5))
plt.plot(df["Tour"], df["Lapins"], label="Lapins", marker="o")
plt.plot(df["Tour"], df["Renards"], label="Renards", marker="x")
plt.xlabel("Tour")
plt.ylabel("Population")
plt.title("volution des Lapins et Renards")
plt.legend()
plt.grid(True)
plt.show()
```
